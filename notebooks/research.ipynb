#### 1. Remove Outliers as discover from EDA file 
##### Winsorization:
Winsorization replaces the extreme values with the nearest non-outlier value. You can choose to replace them with the maximum or minimum non-outlier value.

```python
numerical_columns
```

```python
def find_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

outliers_weight = find_outliers(train_data, 'Item_Weight')
outliers_visibility = find_outliers(train_data, 'Item_Visibility')
outliers_mrp = find_outliers(train_data, 'Item_MRP')
outliers_sales = find_outliers(train_data, 'Item_Outlet_Sales')

print("Number of outliers in Item_Weight:", len(outliers_weight))
print("Number of outliers in Item_Visibility:", len(outliers_visibility))
print("Number of outliers in Item_MRP:", len(outliers_mrp))
print("Number of outliers in Item_Outlet_Sales:", len(outliers_sales))
```

```python
def Capping(column_name):
    q1=train_data[column_name].quantile(0.25)
    q3=train_data[column_name].quantile(0.75)

    iqr=q3-q1

    lower_whisker=q1-1.5*iqr
    upper_whisker=q3+1.5*iqr
    train_data[column_name]=np.where(train_data[column_name]>upper_whisker, upper_whisker, train_data[column_name])
    train_data[column_name]=np.where(train_data[column_name]<lower_whisker, lower_whisker, train_data[column_name])


Capping('Item_Visibility')
Capping('Item_Outlet_Sales')

```

```python
outliers_visibility = find_outliers(train_data, 'Item_Visibility')
outliers_sales = find_outliers(train_data, 'Item_Outlet_Sales')

print("Number of outliers in Item_Visibility:", len(outliers_visibility))
print("Number of outliers in Item_Outlet_Sales:", len(outliers_sales))
```
